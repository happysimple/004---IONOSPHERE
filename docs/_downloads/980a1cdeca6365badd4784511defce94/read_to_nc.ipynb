```python
import scipy.io as spio
import numpy as np
import netCDF4 as nc
import os

mypath='E:\\001---论文\\002---Data\\ON2\\000---数据\\';
mydir=os.listdir(mypath);
for i in range(0,np.size(mydir)):
    infile=mypath+mydir[i]
    s = spio.readsav(infile, python_dict=True, verbose=True)

    year = s['year'];
    doy = s['doy'];
    d_lat = s['d_lat'];
    d_lon = s['d_lon'];
    thelat = np.arange(-90,90,1.78218);
    thelon = np.arange(0,360,1.78218);
    on2_grid_smooth = s['on2_grid_smooth'];

    tmp = s['saved_data']
    num_orbits = tmp.size
    num_data_per_orbit = tmp[0][0].size

    lon = np.zeros(shape=(num_orbits,num_data_per_orbit))
    lat = np.zeros(shape=(num_orbits,num_data_per_orbit))
    on2 = np.zeros(shape=(num_orbits,num_data_per_orbit))
    Fdoy = np.zeros(shape=(num_orbits,num_data_per_orbit))
    for j in np.arange(0,num_orbits):
        lon[j,:] = tmp[j][1];
        lat[j,:] = tmp[j][0];
        on2[j,:] = tmp[j][3];
        Fdoy[j,:] = tmp[j][4];

    outfile=mydir[i][0:13]+'.nc';
    fp = nc.Dataset(outfile,'w',format = 'NETCDF4')
    fp.createDimension('thelon',202)  
    fp.createDimension('thelat',101) 
    fp.createDimension('num_orbits',num_orbits) 
    fp.createDimension('num_data_per_orbit',num_data_per_orbit) 
    fp.createVariable('thelon','f',('thelon'))  
    fp.createVariable('thelat','f',('thelat'))  
    fp.createVariable('on2gridsmooth','f',('thelat','thelon')) 
    fp.createVariable('lon','f',('num_orbits','num_data_per_orbit')) 
    fp.createVariable('lat','f',('num_orbits','num_data_per_orbit')) 
    fp.createVariable('on2','f',('num_orbits','num_data_per_orbit')) 
    fp.createVariable('Fdoy','f',('num_orbits','num_data_per_orbit')) 
    fp.variables['thelon'][:] = thelon
    fp.variables['thelat'][:] = thelat
    fp.variables['on2gridsmooth'][:] = on2_grid_smooth
    fp.variables['lon'][:] = lon
    fp.variables['lat'][:] = lat
    fp.variables['on2'][:] = on2
    fp.variables['Fdoy'][:] = Fdoy
    fp.close()
```
